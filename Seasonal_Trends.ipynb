```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
file_path = "backend/processed-data/strike-data.csv"
bird_strike_data = pd.read_csv(file_path)
```

```python
# Convert the INCIDENT_DATE column to datetime to extract months
bird_strike_data['INCIDENT_DATE'] = pd.to_datetime(bird_strike_data['INCIDENT_DATE'], errors='coerce')
# Extract the month from the date
bird_strike_data['MONTH'] = bird_strike_data['INCIDENT_DATE'].dt.month
```

```python
# Count the number of incidents per month
monthly_trends = bird_strike_data['MONTH'].value_counts().sort_index()
monthly_trends

```

```python
# Plot the seasonal trends
plt.figure(figsize=(10, 6))
plt.plot(monthly_trends.index, monthly_trends.values, marker='o')
plt.title('Seasonal Trends of Bird Strikes by Month', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Bird Strikes', fontsize=12)
plt.xticks(ticks=range(1, 13), labels=[
    'January', 'February', 'March', 'April', 'May', 'June', 
    'July', 'August', 'September', 'October', 'November', 'December'], rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
